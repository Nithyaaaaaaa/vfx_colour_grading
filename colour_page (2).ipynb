# **TASK 4: Color Page**

The Color Page is where the real magic happens. It’s easily one of the most important parts of the edit, because the colors set the entire mood for the viewer.

For this task, I used a **node based workflow**, which allowed me to work on different parts of the image like the trees, building, and water independently without affecting what I'd already refined.

## **Nodes**
In DaVinci Resolve, instead of using layers, we use Nodes. Think of nodes as the building blocks of your edit. Each node represents a different operation, such as an effect, a mask, or a color adjustment. They are connected in a flowchart like structure, allowing for intricate control over every visual element.
![Nodes](https://img.sanishtech.com/u/2e770fecd93267be3c57cd4e8d4301ad.png)

## **My Step by Step Node Workflow**

I used a total of **7 nodes** to transform the flat **S-Log3** footage into something cinematic. Here's a breakdown of what each one did:

### **Node 1: The Basics**

Started with exposure & contrast adjustments using the **Color Wheels**, acting as a *weight balancer*.

### **Node 2: The Left Grass**

Used a **Window** to isolate the grass on the left side and adjusted the **HDR Wheels** to make the green look more lush and alive.

### **Node 3: Refining**

Used **ColorSlice** and **Curves** for mid level refinements and smoother lighting transitions.

### **Node 4: The Building**

Targeted the building with a **Window** and used **Gain** & **HDR adjustments** to make the white walls stand out more.

### **Node 5: The Trees**

Boosted overall saturation and vibrancy in the trees using a targeted **Window**.

### **Node 6: The Water**

Applied **Curves** to add a teal tint to the water and used **Effects** for enhanced polish.

### **Node 7: The Final Look**

Used a **Color Space Transform (CST)** to convert the **Sony S-Log3** footage to **Rec.709**, making it display correctly on standard screens.

## **Key Tools & Features I Used**

### **Curves Graph**
The Curves graph provides precise control over tonal adjustments. By manipulating the line on the graph, I could adjust brightness and contrast across specific ranges. As seen in my work, I used custom curves to fine tune the Red, Green, and Blue levels individually for a perfectly balanced image.
![Curves Graph](https://img.sanishtech.com/u/446a82c8f6fc38e0fb08f1cd1ebb7ffa.png)

### **Primaries: Color Wheels**

Main tools for image balancing:

* **Lift**: Controls the shadows & darkest areas.
* **Gamma**: Adjusts midtones (crucial for perceived brightness).
* **Gain**: Affects highlights & brightest parts.
* **Offset**: Shifts brightness & color across the entire image.

![](https://img.sanishtech.com/u/346d7029e62176b0ffe4bbcdc30660bd.png)

## **Technical Monitoring (Scopes)**

These helped ensure accuracy beyond what the eye sees:

* **Waveform**: Shows brightness from left to right (avoids clipped highlights).
![Waveform](https://img.sanishtech.com/u/cf1e4c3e75628cae56871384644af5ae.png)
* **Parade**: Displays RGB channels separately for finer color balancing.
![Parade](https://img.sanishtech.com/u/598f55dd137ed89d7c2ca2c4ea7fbf2b.png)
* **Vectorscope**: Tracks hue & saturation—ideal for skin tones & natural landscapes.
![Vectorscope](https://img.sanishtech.com/u/56813f35f1cc29a0a526398a5f76379a.png)
* **Histogram**: Graph showing distribution of light & dark values across the frame.
![Histogram](https://img.sanishtech.com/u/875c25f477f8c56c38208414b717f2d2.png)
